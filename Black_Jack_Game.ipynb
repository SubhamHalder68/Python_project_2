### 1.Global varriable settings
 random module import
 declare variables to store suits, suits_shape,ranks,values
 and boolean value playing decalre

```python
import random

suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
suits_shape = {"Spades":"\u2664", "Hearts":"\u2661", "Clubs": "\u2667", "Diamonds": "\u2662"}
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 'Nine':9, 'Ten':10, 'Jack':10,
         'Queen':10, 'King':10, 'Ace':11}

playing = True
```

### 2.Card class
Card objects will be store suit and rank of card and using __str__ print card value

```python
class Card:

    def __init__(self,suit,rank):
        self.suit = suit
        self.rank = rank
        
    def __str__(self):
        return self.rank + ' of ' + suits_shape[self.suit]
```

### 3.Deck class
52 card objects create,shuffle the cards and deal cards to the player and dealer

```python
class Deck:
    
    def __init__(self):
        self.deck = []  # start with an empty list
        for suit in suits:
            for rank in ranks:
                self.deck.append(Card(suit,rank))  # build Card objects and add them to the list
    
    def __str__(self):
        deck_comp = ''  # start with an empty string
        for card in self.deck:
            deck_comp += '\n '+card.__str__() # add each Card object's print string
        return 'The deck has:' + deck_comp

    def shuffle(self):
        random.shuffle(self.deck)
        
    def deal(self):
        single_card = self.deck.pop()
        return single_card
```

### 4.Hand Class
Hold the card object that deal from deck , adjust ace value and adjust player is busted or not

```python
class Hand:
    
    def __init__(self):
        self.cards = []  # start with an empty list as we did in the Deck class
        self.value = 0   # start with zero value
        self.aces = 0  # add an attribute to keep track of aces
        self.flag=0    #add this to check player is busted or not
    def add_card(self,card):
        self.cards.append(card)
        self.value += values[card.rank]
        if card.rank == 'Ace':
            self.aces += 1  # add to self.aces
    
    def adjust_for_ace(self):
        while self.value > 21 and self.aces:
            self.value -= 10
            self.aces -= 1 
    def adjust_for_player_bust(self): #add player to bust
        self.flag=1
```

### 5.Chips class
To keep track of a Player's starting chips, bets, and ongoing winnings

```python
class Chips:
    
    def __init__(self):
        self.total = 100  # This can be set to a default value or supplied by a user input
        self.bet = 0
        
    def win_bet(self):
        self.total += self.bet
    
    def lose_bet(self):
        self.total -= self.bet
```

### 6.take_bet function
Take input the players bets 

```python
def take_bet(chips,player_id):
    
    while True:
        try:
            chips.bet = int(input(f'How many player {player_id} want to bet?'))#must be integer and within 100 
        except ValueError:
            print('Sorry, a bet must be an integer!')#int error handling
        else:
            if chips.bet > chips.total:
                print("Sorry, your bet can't exceed",chips.total)#bet>100(the default value)
            else:
                break
```

### 7.Hit_Stand functions
two functions:
1.hit_or_stand :for player want to hit or stand
2.hit:deal card and adjust ace call

```python
def hit_or_stand(deck,hand,player_id):
    global playing  # to control an upcoming while loop
    
    while True:
        x = input('player '+str(player_id)+" Would you like to Hit or Stand?[h,s]")
        
        if x[0].lower() == 'h':
            hit(deck,hand)  # hit() function defined above

        elif x[0].lower() == 's':
            print("Player "+str(player_id)+ " stands.")
            playing = False

        else:
            print("Sorry, please try again.")
            continue
        break
def hit(deck,hand):
    hand.add_card(deck.deal())
    hand.adjust_for_ace()
```

```python
#To show player hand cards
def player_card_show(player,player_id):
    print(f"\nPlayer {player_id}  Hands cards=>:",*player.cards, sep='\n ')
#To show only card of dealer
def dealer_half_card_show(dealer):
    print("\nDealer's Hand cards:=>")
    print(" <card hidden>")
    print('',dealer.cards[1])  
# To dealer all cards show
def dealer_card_show(dealer):
    print("\nDealer's Hand:", *dealer.cards, sep='\n ')
    print("Dealer's Hand =",dealer.value)
```
